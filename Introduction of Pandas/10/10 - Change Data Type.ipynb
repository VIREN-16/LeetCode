```python
import pandas as pd

def changeDatatype(students: pd.DataFrame) -> pd.DataFrame:
    return students.astype({'grade': 'int'})
```

```python
students = pd.DataFrame(
    {
    "student_id": [1, 2],
    "name": ["Ava", "Kate"],
    "age": [6, 15],
    "grade": [73.0, 87.0]
}
)

```

```python
students.dtypes
```

```python
students1 = changeDatatype(students)
changeDatatype(students1)
```

```python
students1.dtypes
```

```python
def changeDatatype(students: pd.DataFrame) -> pd.DataFrame:
    students["grade"] = students["grade"].astype(int)
    return students
```

## OR

```python
def changeDatatype(students: pd.DataFrame) -> pd.DataFrame:
    students["grade"] = students["grade"].map(int)
    return students
```

```python
students.dtypes
```

```python
students1 = changeDatatype(students)
changeDatatype(students)
```

```python
students1.dtypes
```
