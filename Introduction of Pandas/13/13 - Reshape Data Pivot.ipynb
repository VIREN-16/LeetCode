```python
import pandas as pd

data = {
    "city": [
        "Jacksonville", "Jacksonville", "Jacksonville", "Jacksonville", "Jacksonville",
        "ElPaso", "ElPaso", "ElPaso", "ElPaso", "ElPaso"
    ],
    "month": [
        "January", "February", "March", "April", "May",
        "January", "February", "March", "April", "May"
    ],
    "temperature": [13, 23, 38, 5, 34, 20, 6, 26, 2, 43]
}

weather = pd.DataFrame(data)
weather
```

```python
def pivotTable(weather: pd.DataFrame) -> pd.DataFrame:
    df = weather.pivot(index='month', columns='city', values='temperature')
    df = df.reset_index()          # bring month as column
    df.columns.name = None         # remove "city" label
    return df
```

```python
pivotTable(weather)
```

```python
def pivotTable(weather: pd.DataFrame) -> pd.DataFrame:
    return weather.pivot(index='month', columns='city', values='temperature') \
                  .rename_axis(None, axis=1) \
                  .reset_index()
```
