```python
import pandas as pd
```

```python
def dropMissingData(students: pd.DataFrame) -> pd.DataFrame:
    return students.dropna(subset='name')
```

```python
# 🔹 Create DataFrame
students = pd.DataFrame(
    {
    "student_id": [32, 217, 779, 849],
    "name": ["Piper", None, "Georgia", "Willow"],
    "age": [5, 19, 20, 14]
}
)

students 
```

```python
# 🔹 Apply function
dropMissingData(students)
```
